# Introduction :

Superconductors are the materials which show zero resistance to the flow of current and no loss of energy takes place. There are two types of them -Conventional and Unconventional Superconductors.

Major cause of supercondivity is known to be formation of <b>cooper pairs</b> which is an electron - electron (each being a fermion with half-integral spin) pair which together forms a boson. Bosons can all occupy a same fermi - energy level .

Now in Conventional Superconductors this cooper pair formation is mediated by attraction due to the lattice vibrations also called electron-phonon interaction. And conventional superconductors are all satisfied by the BCS theory.They general need to be cooled down so much and have a very very low <b>critical temperature</b> (the temperature below which a material behave as a superconductor.

Unconventional Superconductors can operate at much higher critical temperatures which makes them more cost effective and efficient.Their superconductivity can't be explained well by conventional BCS- theory.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore', category=FutureWarning, module='seaborn')
```

# Viewing the Dataset :

```python
df=pd.read_csv('featurized.csv')
pd.set_option('display.max_columns', None)
df
```

## Plotting Critical temperature Vs Category of Superconductors:

As we can see in the below two plots of Critical temperature vs Category , the <b>cuperates</b> along with <b>iron based</b> and <b>bismuthates</b>   Superconductors have a greater critical temperature and show High Temperature Superconductivity. Majority of them are <b>Cuperates</b>.

```python
plt.figure(figsize=(12,6))
plt.title('Histogram of Tc by Category')
sns.histplot(data=df , x='Tc' , hue='category' , multiple='stack')
plt.xlabel('Tc (Critical Temperature in K)')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(12,6))
plt.title('Histogram of Tc by Category' )
sns.histplot(data=df , x='Tc' , hue='category' , element='poly')
plt.xlabel('Tc (Critical Temperature in K)')
plt.ylabel('Count')
plt.show()
```

# Filtering Out Cuperates from the Data set:

```python
cuprates=df.loc[df['category'] == 'Cuprates']
pd.set_option('display.max_columns', None)
cuprates
```

# Plot of Tc and Mean Electonegativity of cuperates :

<p> As we can see in the scatter plot below , maximum cuperates have an mean electronegativity range of roughly <b>1.5 - 2.0</b> and <b>2.25 - 2.75</b></p> with peak critical temperatures being shown around <b> Mean electronegativity of 2.5</b>

```python
plt.figure(figsize=(12,6))
plt.title('Scatter Plot of Tc by Mean Electronegativity' )
sns.scatterplot(x='MagpieData mean Electronegativity',y='Tc',data=cuprates)
plt.xlabel(' Mean electronegativity')
plt.ylabel('Critical Temp.(Tc)')
plt.show()
```

## Plot of Tc vs No. of elements in the composition of Cuperates :

<p>As we can infer from the plot below , Cuperates which contains higher number of elements has a higher critical temperature with a limit of 8 elements , We can see the sharp slope when no. of elemnts is increased from 7 to 8 and critical temperature peaks at max. 8 elements.</p>

```python
plt.figure(figsize=(12,6))
plt.title('Line Plot of Tc by No. of Elements' )
sns.lineplot(x='0-norm',y='Tc',data=cuprates , color='black')
plt.xlabel('No. of elements')
plt.ylabel('Critical Temp.(Tc)')
plt.show()
```

```python
dvalence=cuprates.loc[df['avg f valence electrons'] ==0]
fvalence=cuprates.loc[df['avg f valence electrons'] >0]
```

# Plot of Tc vs Average d-valence electrons/Average f-valence electrons :

<p>Looking up the Scatter plots below , we can see that the cuperates which have <b>f-valence electrons</b> have a higher critical temperature upto 140 K  and most of cuperates with f-valence electrons lie in the average  valence f-electrons range of <b>0 - 2</b> with peak Tc at <b> avg. of 1.</b></p>

<p> Whereas for cuperates with valence d electrons the range of avg. valence d-electrons is between <b> 1 - 5</b> with a peak value of Tc being observed at avg of 4.</p>

```python
plt.figure(figsize=(12,6))
plt.title('Scatter Plot of Tc by Average d-valence Electrons' )
sns.scatterplot(x='avg d valence electrons',y='Tc',data=dvalence ,color=['teal'])
plt.xlabel('Avg. d-Valence electrons')
plt.ylabel('Critical Temp.(Tc)')
plt.show()

plt.figure(figsize=(12,6))
plt.title('Scatter Plot of Tc by Average f-valence Electrons' )
sns.scatterplot(x='avg f valence electrons',y='Tc',data=fvalence ,color=['pink'])
plt.xlabel('Average f-valence electrons')
plt.ylabel('Critical Temp.(Tc)')
plt.show()
```

# Summing Up :

We can conclude from the basic analysis above , that cuperates is the class forming High Temperature superconductors , and more the no. of elements in it with a mean electronegativity around 2.5 shows peak in the critical temperature. Also the ones with valence f-electrons show higher critical temperature than the ones with valence d-electrons.

Cuperates form a very important class of High Temperature Superconductors (HTSc). Advantages of HTSc is that they operate at higher critical temperature , as of now it is much cheaper to use liquid nitrogen for cooling them , than liquid helium used to cool conventional superconductors. Also they can operate at high critical Magnetic Field , which allows them to be used as superconducting magnets. They do not break the meissener effect even at a higher magnetic fields , which is required when working with supermagnets.
